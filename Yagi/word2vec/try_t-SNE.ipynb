# t-SNEに挑戦
---
sciket-learnを使ってt-SNEを試してみる．

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from gensim.models import word2vec
from sklearn.manifold import TSNE

plt.rcParams['font.family'] = 'Osaka'
```

### まずはword2vec

```python
data = word2vec.Text8Corpus("basket.txt")
model = word2vec.Word2Vec(data, size=50)
```

```python
category = list(model.vocab.keys())
X = np.vstack([model[category[c]] for c in range(len(category))])
```

### t-SNEを学習

```python
model2 = TSNE(n_components=2)
Z = model2.fit_transform(X)
```

### いざ可視化

```python
plt.scatter(Z[:,0], Z[:,1])
```

ダマになってる...。うれしくない展開  
とりあえず、各カテゴリーの位置を確認

```python
fig = plt.figure(figsize=(80, 80))
plt.scatter(Z[:,0], Z[:,1])
for i in range(len(category)):
    x = Z[i, 0]
    y = Z[i, 1]
    plt.text(x, y, category[i])
```
