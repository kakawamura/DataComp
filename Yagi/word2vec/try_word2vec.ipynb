# word2vecに挑戦
---
gensimを使ってword2vecを試してみる.

## モデルの学習

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from gensim.models import word2vec
```

データの読み込み,モデルの学習を行う．

```python
data = word2vec.Text8Corpus("basket.txt")
model = word2vec.Word2Vec(data, size=50, iter=50)
```

学習された語彙（商品カテゴリ）を表示

```python
vocab = list(model.vocab.keys())
print(vocab)
```

## 結果を確認

各カテゴリに類似するカテゴリを降順に表示(数値は類似度）

```python
for v in vocab:
    print("\n# " +v)
    out = model.most_similar(positive=v)
    for x in out:
        print(x[0], x[1])
```

類似度の高い関係があるもののみ表示（類似度0.85以上)

```python
for v in vocab:
    out = model.most_similar(positive=v)
    if out[0][1] > 0.85:
        print("\n# " +v)
        for x in out:
            print(x[0], x[1])
```

## ベクトル演算に挑戦

「"スーツジャケット" - "テーラードジャケット" + "パンツ" =???　」を求めてみる．

```python
out = model.most_similar(positive=["スーツジャケット", "パンツ"], negative=["テーラードジャケット"])
for x in out:
    print(x[0], x[1])
```

”スーツジャケット”　 -  ”テーラードジャケット”　+ ”パンツ" = ”スーツパンツ”になった！ （それっぽい）
