結論，よくわかんなかった． きちんとしたテストを考える必要があるのだろうか...。

```python
import numpy as np
from gensim.models import word2vec

plt.rcParams['font.family'] = 'Osaka'
```

```python
data = word2vec.Text8Corpus("basket.txt")
def run_word2vec(SENTENCES=data, SG=0, SIZE=50, WINDOW=5,
                 ALPHA=0.025, MIN_ALPHA=0.0001, MIN_COUNT=5, NEGATIVE=5,
                ITER=5):
    from gensim.models import word2vec
    model = word2vec.Word2Vec(sentences=SENTENCES, sg=SG, size=SIZE, window=WINDOW, 
                              alpha=ALPHA, min_alpha=MIN_ALPHA, min_count=MIN_COUNT, negative=NEGATIVE,
                             iter=ITER, )
    
    vocab = list(model.vocab.keys())
    
    print("\n# スーツジャケット - テーラードジャケット + パンツ = ")
    out = model.most_similar(positive=["スーツジャケット", "パンツ"], negative=["テーラードジャケット"])
    for x in out:
        print(x[0], x[1])
        
    print("\n# デニムパンツ - パンツ + スカート = ")
    out = model.most_similar(positive=["デニムパンツ", "スカート"], negative=["パンツ"])
    for x in out:
        print(x[0], x[1])
        
        
    print("\n# サングラス - メガネ + キャップ = ")
    out = model.most_similar(positive=[ "サングラス","キャップ"], negative=["メガネ"])
    for x in out:
        print(x[0], x[1])
    
    for v in vocab:
        out = model.most_similar(positive=v)
        if out[0][1] > 0.0:
            print("\n# " +v)
            for x in out:
                print(x[0], x[1]) 
                
```

# モデルの比較(CBOW, Skip-gram)

```python
run_word2vec() # CBOW
```

```python
run_word2vec(SG=1) # Skip gram
```

Skip-gramで良さそう．
