```python
% matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
import os
```

```python
def idx2color(idx):
    D = {0:'#e41a1c', 1:'#377eb8', 2:'#4daf4a', 3:'#ffff33', 4:'#f781bf'}
    return D[idx]

def f(path, K):
    files = os.listdir(path)
    col_name = np.loadtxt(path + 'itemdist_columns.txt', delimiter=',', dtype=int)
    
    itemdists = []
    for i in range(13):
        tmp = np.loadtxt(path + files[i], delimiter=',')
        itemdists.append(tmp)
        
    colors = list(map(idx2color, col_name))
    
    fig, axes = plt.subplots(13, K, figsize=(30, 30))
    x = np.arange(5)
    
    for i in range(13):
        for j in range(K):
            axes[i, j].bar(x,itemdists[i][j], color=colors, alpha=0.7)
            if i==0:
                axes[i, j].set_title('cluster%d' %j)
            if j==0:
                axes[i, j].set_ylabel('month%d'  %i)
            axes[i, j].set_xticks(x+0.5)
            axes[i, j].set_xticklabels(colors)
```
