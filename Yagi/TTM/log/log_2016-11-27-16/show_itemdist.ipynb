```python
% matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
import os
```

```python
def idx2color(idx):
    D = {0:'red', 1:'darkorange', 2:'yellow', 3:'lime', 4:'cyan', 5:'dodgerblue', 6:'blue', 7:'blueviolet', 8:'magenta', 9:'deeppink'}
    return D[idx]

def f(path, K):
    files = os.listdir(path)
    col_name = np.loadtxt(path + 'itemdist_columns.txt', delimiter=',', dtype=int)
    
    itemdists = []
    for i in range(12):
        tmp = np.loadtxt(path + files[i], delimiter=',')
        itemdists.append(tmp)
        
    colors = list(map(idx2color, col_name))
    
    fig, axes = plt.subplots(12, K, figsize=(30, 30))
    x = np.arange(10)
    
    for i in range(12):
        for j in range(K):
            axes[i, j].bar(x,itemdists[i][j], color=colors, alpha=0.7)
            if i==0:
                axes[i, j].set_title('cluster%d' %j)
            if j==0:
                axes[i, j].set_ylabel('month%d'  %i)
            axes[i, j].set_xticks(x+0.5)
            axes[i, j].set_xticklabels(colors)
```
